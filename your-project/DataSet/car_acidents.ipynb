```python
import pandas as pd
```

```python
car_accidents = pd.read_csv("2018_accidents_causes_gu_bcn_.csv")

car_accidents.head()
```

```python
car_accidents = pd.concat([car_accidents, pd.get_dummies(car_accidents ["Descripcio_causa_mediata"])], axis=1)
car_accidents.head()
```

```python
sum_val = car_accidents.groupby('Nom_districte', as_index=True).agg({'Alcoholèmia':'sum',
                                                                    "Calçada en mal estat": "sum",
                                                                    "Drogues o medicaments": "sum",
                                                                    "Estat de la senyalització":"sum",
                                                                    "Excés de velocitat o inadequada": "sum",
                                                                    "Factors meteorològics":"sum",
                                                                    "No hi ha causa mediata": "sum",
                                                                    "Objectes o animals a la calçada": "sum",
                                                                    "Numero_expedient":"count"})
#sum_nal = emigrants.groupby("Nom_districte", as_index=True).agg({"Espanyola": "count"})
print(sum_val)
#print(sum_nal)
```

```python
pd.DataFrame.transpose(sum_val)
```

```python
# Exporting

car_acidents_values = sum_val.to_csv("2018_emigrants.csv")
```
